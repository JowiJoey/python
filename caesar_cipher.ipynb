```python
import string

def ROT(input_string, shift, encrypt_decrypt):
    ''' Caesar Cipher (encrypt / decrypt shifting the alphabet)
    input_string = string type
    shift = integer type
    encrypt_decrypt = 0 (encrypt) or 1 (decrypt) '''

    if type(input_string) is not str:
        raise ValueError('input argument index 0 must be string type')
    elif type(shift) is not int:
        raise ValueError('input argument index 1 must be integer type')
    elif encrypt_decrypt not in (0,1):
        raise ValueError('input argument index 2 be 0 or 1')

    output_string = ''
    for char in input_string:
        if char in string.ascii_lowercase:
            char_set = string.ascii_lowercase
            output_string += get_sister(char, char_set, shift, encrypt_decrypt)
        elif char in string.ascii_uppercase:
            char_set = string.ascii_uppercase
            output_string += get_sister(char, char_set, shift, encrypt_decrypt)
        else:
            output_string += char
        return output_string

def get_sister(char, char_set, shift, encrypt_decrypt):
    ''' Retrieve shifted character '''
    ## Verify given character is include in given set
    if char not in char_set: raise ValueError('char not in char_set')
    char_set_len = len(char_set)
    shift = abs(char_set_len * encrypt_decrypt - shift)
    return char_set[(char_set.find(char) + shift) %char_set_len]
```

```python
import base64

## make string to bytes object
string_to_encode = str.encode('Encode me!')

## do a base64 encoding
encode = base64.b64encode(string_to_encode)

## should output b'RW5jb2RlIG1lIQ==' where the b is for bytes
print(encode)
```

```python
## do base64 decoding
data = base64.b64decode(encode)

## make the bytes object back to a string
string_to_decode = data.decode('utf-8')

## should give you back the original message 'Encode me!'
print(string_to_decode)
```

```python
import binascii

## make the string to a byte object
string_to_hex = str.encode('Hex me!')

## hexlify
encoded = binascii.hexlify(string_to_hex)

#should output b'486578206d6521' where b is for bytes
print(encoded)
```

```python
## un-hexlify
data = binascii.unhexlify(encoded)

## make the bytes object back to a string
string_to_unhex = data.decode('utf-8')

## should give us back the original 'Hex me!'
print(string_to_unhex)
```
